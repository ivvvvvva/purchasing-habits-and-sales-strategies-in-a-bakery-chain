# Racun

Prebacivanje podataka u listu radi lakse manipulacije

```python
p = []
for index, row in tabela.iterrows():
    i = [x for x in row.values]
    p.append(i)
```

```python
p[:20]
```

Leksikografko sortiranje pre grupisanje racuna po mesecima

```python
p.sort(key=lambda x:x[1].split('-')[1])
```

```python
p[:10]
```

```python
for element in p:
    element[1] = element[1].split('-')[1]
```

```python
p
```

```python
from itertools import groupby
```

Smestanje podataka u mapu gde je mesec kljuc

```python
grupisano = {k:list(g) for k,g in groupby(p, key=lambda x: x[1])}
```

```python
grupisano['DEC']
```

Sumiranje zarade po mesecima

```python
zarada = []
for key, val in grupisano.items():
    suma = 0
    for x in val:
        suma += x[4] * x[5]
    zarada.append((suma, key))
```

```python
zarada
```

Grupisanje po delu godine

```python
suma_p = 0
suma_s = 0
suma_k = 0
for x in zarada:
    if x[1] in pocetak_godine:
        suma_p += x[0]
    elif x[1] in sredina_godine:
        suma_s += x[0]
    else:
        suma_k += x[0]
```
