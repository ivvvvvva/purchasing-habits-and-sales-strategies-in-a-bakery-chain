# Racun

Prebacivanje u listu

```python
p = []
for index, row in tabela.iterrows():
    i = [x for x in row.values]
    p.append(i)
```

```python
p
```

```python
from itertools import groupby
```

Grupisanje po id-u zaposlenog

```python
grupisano = {k:list(g) for k,g in groupby(p, key=lambda x: x[0])}
```

```python
grupisano
```

Racunjanje ukupne zarade i ispis rezultata

```python
zarade = []
for emp, val in grupisano.items():
    suma = 0
    ime = val[0][2]
    prezime = val[0][1]
    for lista in val:
        suma += lista[4] * lista[5]
    print(f'Prodavac {emp} je zaradio: {suma}')
    t = (emp, suma, ime, prezime)
    zarade.append(t)
```

```python
zarade.sort(key= lambda x: x[1])
```
