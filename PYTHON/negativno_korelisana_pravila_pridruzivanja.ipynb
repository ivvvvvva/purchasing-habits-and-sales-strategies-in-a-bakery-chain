Racunanje podrske za svaku stavku posebno

```python
def izracunaj_podrsku(transakcije, stavka):
    broj_transakcija = len(transakcije)
    broj_pojava = sum([1 for transakcija in transakcije if stavka in transakcija])
    podrska = broj_pojava / broj_transakcija
    return podrska
```

```python
def izracunaj_podrsku_za_sve_stavke(transakcije):
    sve_stavke = set(item for transakcija in transakcije for item in transakcija)
    podrska_po_stavci = {}

    for stavka in sve_stavke:
        podrska = izracunaj_podrsku(transakcije, stavka)
        podrska_po_stavci[stavka] = podrska

    return podrska_po_stavci
```

```python
podrska_po_stavci = izracunaj_podrsku_za_sve_stavke(transakcije)
```

Ispis rezultata

```python
for stavka, podrska in podrska_po_stavci.items():
    print(f"Podrska za {stavka}: {podrska}")
```

Pravila pridruzivanja u IBM SPSS Modeleru

```python
pravila_pridruzivanja = pd.read_csv('spss_pravila.csv', delimiter='\t')
```

```python
pravila_pridruzivanja
```

```python
skupovi_stavki = []
for index, row in pravila_pridruzivanja.iterrows():
    i = [x for x in row.values]
    skupovi_stavki.append([i[0], i[1]])
```

```python
skupovi_stavki
```

Uslov koji mora vaziti da bi neko pravilo bilo negativno korelisano:  

Od stavki koje se zajedno javljaju bar jedna mora biti negativna, a ostale pozitivne.

```python
def uslov(lista_stavki, mapa_podrske, minSup):
    drugo = lista_stavki[1].split()
    if len(drugo) > 1:
        drugo = [x for x in drugo if x != 'and']
    stavke = [lista_stavki[0]]
    for i in drugo:
        stavke.append(i)
    negativna = False
    for stavka in stavke:
        if mapa_podrske[stavka] < minSup:
            negativna = True
    return negativna
```

U nastravku je ispis negativno korelisanih pravila pridruzivanja. 

```python
for pravilo in skupovi_stavki:
    if uslov(pravilo, podrska_po_stavci, 0.07):
        print("Pravilo " + "{" + pravilo[1] + "}" + " -> {" + pravilo[0] + "} je negativno korelisano.")
```
